```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[2,4,6,8]


plt.figure(figsize=(8,5))
plt.plot(x,y,label='first line',color="b")
plt.plot([3,6,9],[2,3,4],label='secound line',color="r",linestyle='dotted')
plt.xlabel("x-cordinates")
plt.ylabel('y-cordinates')
plt.title("LINE GRAPH")
plt.legend()
plt.show
```

# SCATTER GRAPH

```python
x=[1,2,3,4]
y=[2,4,6,8]


plt.figure(figsize=(8,5))
plt.scatter(x,y,label='first line',color="b",marker='d',s=200)
plt.scatter([3,6,9],[2,3,4],label='secound line',color="r")
plt.xlabel("x-cordinates")
plt.ylabel('y-cordinates')
plt.title("LINE GRAPH")
plt.legend()
plt.show
```

# BAR GRAPH

```python
ind_over = [1,2,3,4,5,6,7,8]
ind_runs = [20,30,25,20,35,10,8,9]

plt.bar(ind_over,ind_runs,color="g")
plt.xlabel("ind_overs")
plt.ylabel('ind_runs')
plt.title("BAR GRAPH")
plt.legend()
plt.show
```
