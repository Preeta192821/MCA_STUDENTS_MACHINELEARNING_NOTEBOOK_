# DECISION TREE REGRESSOR

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv(r"D:\Datasets\Position_Salaries.csv")
df.head(10)
```

```python
plt.scatter(df.Level,df.Salary)
plt.xlabel("LEVEL")
plt.ylabel("SALARY")
```

```python
df.head()
```

```python
x = df.iloc[:,1:2].values

x
```

```python
y = df.Salary
y
```

```python
from sklearn.tree import DecisionTreeRegressor

clf = DecisionTreeRegressor()
clf.fit(x,y)
```

```python
clf.predict([[5.01]])
```

```python
df.head()
```

# VISUALIZATION

```python
import numpy as np
x_grid = np.arange(min(x),max(x),0.01) # this is in vector form
x_grid = x_grid.reshape(len(x_grid),1) # we change vector into matrix
x_grid
```

```python
plt.scatter(x,y,color='r')
plt.plot(x_grid,clf.predict(x_grid))
#plt.plot(x,clf.predict(x))
```

```python
plt.scatter(x,y,color='r')
#plt.plot(x_grid,clf.predict(x_grid))
plt.plot(x,clf.predict(x))
```

```python
clf.score(x,y)
```
